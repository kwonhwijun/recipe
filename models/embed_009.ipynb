```python
def oracleTopd(query):
    import oracledb as od
    import pandas as pd
    import config

    # db connection
    od.init_oracle_client(lib_dir=r"C:\Program Files\Oracle\instantclient_21_12")
    conn = od.connect(user = config.DB_CONFIG['user'], password = config .DB_CONFIG['password'], dsn = config.DB_CONFIG['dsn'])
    exe = conn.cursor()
    exe.execute(query)

    row = exe.fetchall() # row 불러오기

    column_name = exe.description # column 불러오기
    columns=[]

    for i in column_name:
        columns.append(i[0])

    # row, column을 pandas DataFrame으로 나타내기
    result = pd.DataFrame(row, columns=columns)

    # dtype clob을 string으로 변환
    for clob in result.columns:
        if result[clob].dtype == object:
            result[clob] = result[clob].astype("string")

    conn.close()

    return result
```

```python
import pandas as pd
data = pd.read_csv(r'data\RecipeData.csv')
recipe_ingre_pd = data[['recipe_title', 'recipe_ingredients']]
recipe_ingre_pd
```

```python
def combined_filter(data):
    def clean_ingredients(ingredient): # 이상한 문자열 저장방식 지움
        return ingredient.replace('\\ufeff', '').replace('\\u200b', '')

    data['recipe_ingredients'] = clean_ingredients(data['recipe_ingredients'])
    
    def not_empty_ingredients(row):
        return data['recipe_ingredients'].strip() != '{}' # 결측치 제거

    return not_empty_ingredients(data)

recipe_ingre_pd_pre = recipe_ingre_pd[recipe_ingre_pd.apply(combined_filter, axis=1)]
recipe_ingre_pd_pre
```

```python
for i in range(1, 25):
    recipe_ingre_pd_pre.loc[:, f'ingredient{i}'] = None
    recipe_ingre_pd_pre.loc[:, f'quantity{i}'] = None
    recipe_ingre_pd_pre.loc[:, f'unit{i}'] = None
```

```python
from tqdm import tqdm
import ast
import re
import json

non_matching_items = {}

for idx, row in tqdm(recipe_ingre_pd_pre.iterrows(), total = recipe_ingre_pd_pre.shape[0]):
    ingredient_count = 1
    if idx < 2000:
        ingredients_dict = ast.literal_eval(row["recipe_ingredients"]) # SyntaxError: unterminated string literal (detected at line 1)
        for items in ingredients_dict.values():
            for item in items :
                match = re.match(r'([가-힣]+(\([가-힣]+\))?|\(\w+or\w+\)|\([가-힣]+\)[가-힣]+|\d+[가-힣]+)([\d.+/~-]*)([가-힣a-zA-Z]+|약간|조금)?', item)
                if match:
                    ingredient, _, quantity, unit = match.groups()

                    recipe_ingre_pd_pre.loc[idx, f'ingredient{ingredient_count}'] = ingredient
                    recipe_ingre_pd_pre.loc[idx, f'quantity{ingredient_count}'] = quantity
                    recipe_ingre_pd_pre.loc[idx, f'unit{ingredient_count}'] = unit

                    ingredient_count += 1
                else:
                    non_matching_items[idx] = item

for idx, item in non_matching_items.items():
    print(f'Row {idx}: {item}')
```
