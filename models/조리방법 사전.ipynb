```python
from konlpy.tag import Okt
okt = Okt()
from tqdm import tqdm
```

```python
def oracleTopd(query):    
    import oracledb as od
    import pandas as pd

    # db connection
    od.init_oracle_client(lib_dir=r"C:\Program Files\Oracle\instantclient_21_12")
    conn = od.connect(user = 'admin', password = 'INISW2inisw2',
                      dsn = 'inisw2_high')
    exe = conn.cursor()
    exe.execute(query)
    
    row = exe.fetchall() # row 불러오기

    column_name = exe.description # column 불러오기
    columns=[]
    
    for i in column_name:
        columns.append(i[0])
    
    # row, column을 pandas DataFrame으로 나타내기
    result = pd.DataFrame(row, columns=columns)
    result.rename(mapper=str.lower, axis='columns', inplace=True)
    
    # dtype clob을 string으로 변환
    for clob in result.columns:
        if result[clob].dtype == object:
            result[clob] = result[clob].astype("string")
    
    conn.close()
    
    return result

query = 'select * from recipe_table'
df = oracleTopd(query)
df
```

```python
df1 = df[['recipe_url', 'recipe_title', 'recipe_step']]
```

```python
#recipe_step이 None인 행 출력
df1[df1['recipe_step'].isnull()]
```

```python
#recipe_step이 None인 행 삭제
df2 = df1.dropna(subset=['recipe_step'])
```

```python
recipe_step = df2['recipe_step']
```

```python
recipe_step_30 = recipe_step[:]
```

```python
stems = []
for recipe_step in tqdm(recipe_step_30):
    pos_list = okt.pos(recipe_step)
    verbs = [word for word, pos in pos_list if pos.startswith('Verb')]
    for verb in verbs:
        stem = okt.morphs(verb, stem = True)
        stems.append(stem[0])
```

```python
stems
```

```python
unique_stems = list(set(stems))
unique_stems
```

### 동사 어간 빈도수 추출

```python
from collections import Counter

word_freq = Counter(stems)  # 데이터에서 각 단어의 등장 횟수 계산

for word, freq in word_freq.items():
    print(f"'{word}'의 등장 횟수: {freq}번")
```

```python
word_freq = Counter(stems)  # 데이터에서 각 단어의 등장 횟수 계산

# 데이터 프레임 생성
df = pd.DataFrame(list(word_freq.items()), columns=['단어', '빈도'])
df_verb = df.sort_values(by='빈도', ascending=False)
df_verb
```

```python
df_verb_recipe = df_verb
df_verb_recipe = df_verb_recipe[:1]
df_verb_recipe2 = df_verb_recipe
df_verb_recipe2
```

```python
for stem in unique_stems:
    df_verb_recipe2[stem] = None
del df_verb_recipe2['단어']
del df_verb_recipe2['빈도']
```

```python
df_verb_recipe3 = df_verb_recipe2[:0]
df_verb_recipe3
```

```python
for recipe_step in tqdm(recipe_step_30):
    recipe_step_stems = []
    pos_list = okt.pos(recipe_step)
    verbs = [word for word, pos in pos_list if pos.startswith('Verb')]
    
    for verb in verbs:
        stem = okt.morphs(verb, stem = True)
        recipe_step_stems.append(stem[0])
        
    word_counts = Counter(recipe_step_stems)
                             
    for uni_stem in unique_stems:
        if word_counts[uni_stem] == 0:
            word_counts.update({uni_stem : 0})
            
    # 새로운 행을 생성하여 데이터프레임에 추가
    word_counts_row = pd.DataFrame([word_counts])
    df_verb_recipe3 = pd.concat([df_verb_recipe3, word_counts_row], ignore_index=True)
df_verb_recipe3
```
