```python
import re
def extract_verbs(sentence):
    pos = okt.pos(sentence)
    verbs = [word for word, tag in pos if tag.startswith('V')]
    verb_list = []
    for verb in verbs:
        stem = okt.morphs(verb, stem = True)
        verb_list.append(stem[0])
    return verb_list

step_list = []
for step in tqdm(recipe_step_30):
    has_number = lambda s: bool(re.search(r'\d', s))
    
    # 숫자가 없는 원소를 필터링하여 새로운 리스트 생성
    filtered_data = list(filter(lambda x: has_number(x),  step.split('#')))
    
    # 추출된 동사 리스트로 저장
    verbs_per_element = [extract_verbs(sentence) for sentence in filtered_data]
    step_list.append(verbs_per_element)
```

```python
import pandas as pd

# 데이터 준비
data = {'step': step_list}

# 데이터프레임 생성
df = pd.DataFrame(data)
df
```
