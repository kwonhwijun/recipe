```python
import csv

with open('./data/change2.txt', "r", encoding='utf-8') as f:
    reader = csv.reader(f, delimiter=" ")
    d = {}
    for row in reader:
        # 첫 번째 칼럼을 키로 사용
        key = row[0]
        # 두 번째 칼럼을 value로 사용
        value = row[1]
        # 두 번째 칼럼을 키로 하는 딕셔너리를 생성
        if key not in d:
            d[key] = {}
        # value를 해당 딕셔너리에 추가
        d[key][value] = 1
```

```python
change_path = './data/change2.txt'
dic = {}
with open(change_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines:
        food, unit, gram = line.split(" ")
        if food not in dic:
            dic[food] = [[unit,gram]]
        else: dic[food].append([unit, gram])
print(dic)
```

```python
def find_units_for_ingredient(dataframe, ingredient_name):
    units = set()  # 중복을 허용하지 않는 단위를 저장하기 위한 집합(set)

    for i in range(1, 4):
        ingredient_col = f'ingredient{i}'
        unit_col = f'unit{i}'

        ingredients = dataframe[ingredient_col]
        units_col = dataframe[unit_col]

        for ingredient, unit in zip(ingredients, units_col):
            if ingredient == ingredient_name and not pd.isna(unit):
                units.add(unit)


    return units - set(absolute_conversion_factors.keys())

find_units_for_ingredient(final, '돼지고기')
```
