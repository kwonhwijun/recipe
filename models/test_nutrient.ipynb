```python
import pandas as pd

query = 'SELECT * FROM NUTRIENT_DATA_TABLE'
exe.execute(query)
data = exe.fetchall()

col_data = exe.description # col_data : "('NUTRIENT', <DbType DB_TYPE_VARCHAR>, 300, 300, None, None, True)"
col_name = [i[0] for i in col_data] # col_name : ['NUTRIENT',  '수분',  '에너지(Kcal)',  '에너지(kJ)']
df = pd.DataFrame(data, columns = col_name)
df.columns = df.columns.map(lambda x: x.lower()) # 소문자로 변환하기
```
